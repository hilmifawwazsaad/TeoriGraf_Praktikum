**Largest Monotonically Increasing Subsequence**
Picture1.png

```python
def lis(arr):
    def _lis(n):
        nonlocal maximum

        if n == 1:
            return [arr[0]]

        maxEndingHere = [arr[n-1]]

        for i in range(1, n):
            res = _lis(i)
            if arr[i-1] < arr[n-1] and len(res) + 1 > len(maxEndingHere):
                maxEndingHere = res + [arr[n-1]]

        maximum = max(maximum, maxEndingHere, key=len)

        print(f"{arr[n-1]}: {maxEndingHere}")

        return maxEndingHere

    n = len(arr)
    maximum = [arr[0]]
    _lis(n)
    return maximum

if __name__ == '__main__':
    arr = [4,1,13,7,0,2,8,11,3]
    n = len(arr)

    result = lis(arr)

    print("\nLength of LMIS", arr, " is ", len(result))
    print("LIS Path:", result)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.lis = None

def insert(root, value):
    if root is None:
        return Node(value)

    if value < root.value:
        root.left = insert(root.left, value)
    else:
        root.right = insert(root.right, value)

    return root

def find_lis(node, prev_value):
    if node is None:
        return []

    current_lis = [node.value]
    left_lis = find_lis(node.left, node.value)
    right_lis = find_lis(node.right, node.value)

    if prev_value is None or node.value > prev_value:
        current_lis.extend(max(left_lis, right_lis, key=len))

    node.lis = current_lis

    return current_lis

def largest_monotonically_increasing_subsequence(sequence):
    root = None
    for num in sequence:
        root = insert(root, num)

    lis = find_lis(root, None)
    return max(lis)

# Contoh penggunaan
sequence = [4, 1, 13, 7, 0, 2, 8, 11, 3]
result = largest_monotonically_increasing_subsequence(sequence)

print("Urutan bilangan:", sequence)
print("Largest Monotonically Increasing Subsequence:", result)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def lis(arr):
    def _lis(n):
        nonlocal maximum

        if n == 1:
            return [arr[0]]

        maxEndingHere = [arr[n-1]]

        for i in range(1, n):
            res = _lis(i)
            if arr[i-1] < arr[n-1] and len(res) + 1 > len(maxEndingHere):
                maxEndingHere = res + [arr[n-1]]

        maximum = max(maximum, maxEndingHere, key=len)

        print(f"{arr[n-1]}: {maxEndingHere}")

        return maxEndingHere

    n = len(arr)
    maximum = [arr[0]]
    _lis(n)
    return maximum

def draw_tree(G, node, pos=None, level=1, width=2., vert_gap=0.4, xcenter=0.5):
    if pos is None:
        pos = {node: (xcenter, 1 - level * vert_gap)}
    else:
        pos[node] = (xcenter, 1 - level * vert_gap)

    neighbors = list(G.neighbors(node))
    if len(neighbors) != 0:
        dx = width / 2
        nextx = xcenter - width / 2 - dx / 2
        for neighbor in neighbors:
            nextx += dx
            pos = draw_tree(G, neighbor, pos=pos,
                            level=level + 1, width=dx, xcenter=nextx)

    return pos

if __name__ == '__main__':
    arr = [4, 1, 13, 7, 0, 2, 8, 11, 3]
    n = len(arr)

    result = lis(arr)

    print("Length of lis is", len(result))
    print("LIS Path:", result)

    # Membuat pohon
    G = nx.DiGraph()
    for i in range(1, n + 1):
        G.add_node(f"Node {i}", label=str(arr[i - 1]))

    for i in range(1, n):
        G.add_edge(f"Node {i}", f"Node {i + 1}")

    # Tampilkan visualisasi pohon
    pos = draw_tree(G, "Node 1")
    labels = nx.get_edge_attributes(G, 'label')
    nx.draw(G, pos, labels=labels, with_labels=True, node_size=2000, node_color="skyblue",
            font_size=10, font_color="black", font_weight="bold", arrows=True, connectionstyle='arc3,rad=0.1')
    plt.title("LIS Path Tree")
    plt.show()
```
